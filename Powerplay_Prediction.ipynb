```python
df['byes'] = df['byes'].fillna(0)
df['noballs'] = df['noballs'].fillna(0)
df['wides'] = df['wides'].fillna(0)
df['legbyes'] = df['legbyes'].fillna(0)
df['player_dismissed'] = df['player_dismissed'].fillna('notout')
df['wicket_type'] = df['wicket_type'].fillna('nowicket')
```

```python
df = df.replace(['M Chinnaswamy Stadium',"M.Chinnaswamy Stadium"],"M Chinnaswamy Stadium")
df = df.replace(['MA Chidambaram Stadium, Chepauk',"MA Chidambaram Stadium, Chepauk, Chennai"],"MA Chidambaram Stadium")
df = df.replace(['Punjab Cricket Association IS Bindra Stadium',"Punjab Cricket Association Stadium, Mohali"],"Punjab Cricket Association Stadium, Mohali")
df = df.replace(['Wankhede Stadium',"Wankhede Stadium, Mumbai","Maharashtra Cricket Association Stadium"],"Wankhede Stadium")
df = df.replace(['Rajiv Gandhi International Stadium, Uppal',"Rajiv Gandhi International Stadium"],"Rajiv Gandhi International Stadium")
df = df.replace(['Rajiv Gandhi International Stadium, Uppal',"Rajiv Gandhi International Stadium"],"Rajiv Gandhi International Stadium")
df = df.replace(['Delhi Daredevils'],"Delhi Capitals")
df = df.replace(['Deccan Chargers'],"Sunrisers Hyderabad")
df = df.replace(['Rising Pune Supergiant'],'Rising Pune Supergiants')
df = df.replace(["Punjab Cricket Association IS Bindra Stadium, Mohali","Punjab Cricket Association IS Bindra Stadium","Punjab Cricket Association Stadium, Mohali"],"Punjab Cricket Association Stadium, Mohali")
```
